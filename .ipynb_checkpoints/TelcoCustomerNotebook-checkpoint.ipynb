## Keşifsel Veri Analizi (Exploratory Data Analysis - EDA)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Verisetini yüklüyoruz
df = pd.read_csv("telco.csv")
```

```python
# Verisetinin ilk 5 satırını görüntüleyerek, veriseti hakkında ilk adım olarak fikir sahibi oluyoruz
df.head()
```

```python
# Verisetindeki sütunlar, null olmayan değer sayıları, veri tipleri hakkında genel bilgi 
df.info()
```

```python
# info() fonksiyonuyla da alınabilen bilgileri, aşağıdaki yöntemle, tablo halinde daha okunaklı şekilde gösterebiliriz.

df_info = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes.values,
    "Missing Value Count": df.isnull().sum().values,
    "Missing Value Ratio (%)": np.round(df.isnull().mean().values * 100, 2),
    "Filled Value Count": df.notnull().sum().values
})

df_info = df_info.sort_values("Missing Value Count", ascending=False).reset_index(drop=True)

df_info
```

```python
# Verisetinin sayısal sütunlarının özet istatistiklerini görüntülüyoruz
df.describe().T
```

### Eksik Değerlerin İncelenmesi ve Müdahale

```python
missing_value_count = df.isnull().sum() # Her sütundaki eksik değer sayısını hesaplıyoruz.
missing_value_count = missing_value_count[missing_value_count > 0] # Sadece eksik değeri olan sütunları seçiyoruz

# Eğer hiç eksik değer yoksa hata vermemek için kontrol ediyoruz.
if not missing_value_count.empty:

    missing_value_ratio = missing_value_count / len(df) * 100
    
    # Sadece eksik veri içeren sütunları karşılaştırmak için yeni bir data frame oluşturuyoruz.
    missing_df = pd.DataFrame({
    "Column Name": missing_value_ratio.index,
    "Missing Value Ratio": missing_value_ratio.values
    })
    
    # Oranlara göre sıralıyoruz. (isteğe bağlı, daha düzenli görünüm sağlar)
    missing_df = missing_df.sort_values(by='Missing Value Ratio', ascending=False)

    # Oluşturduğumuz data frame ile eksik veri görselleşitrmesi yapıyoruz.
    plt.figure(figsize=(15,10))
    sns.catplot(x="Column Name", y="Missing Value Ratio", data=missing_df, hue="Column Name", legend=False, kind="bar", palette="RdPu")
    plt.title("Missing Value Ratio")
    plt.xlabel("Missing Value Columns")
    plt.ylabel("% Ratio")
    plt.xticks(rotation=45, ha='right') # Sütun isimleri çakışmasın diye döndür
    plt.tight_layout()
    plt.show()

else:
    print("There is no missing value in the data set.")
```

Churn Reason, Churn Category, Offer, Internet Type sütunlarında boş değerler olduğunu görüyoruz. Bu değişkenler kategorik değişkenlerdir. Eğer sayısal değişkenler olsaydı bir korelasyon analizi yapmamız gerekebilirdi. Doğrusal bir korelasyon olup olmadığını incelememiz gerekirdi.

1. Churn Reason ve Churn Category
Churn Reason ve Churn Category sütunlarındaki boş değer oranı çok fazla gözüküyor. Bu boş değerler, eğer müşterinin terk etmediği yani Churn Label = "No" olan değerlerse, Churn Reason ve Churn Category alanlarının boş değer olması beklenen bir durumdur. Öncelikle bu durumu kontrol etmek gerekir.

Müdahale: Bu sütunlar, Churn Label hedef değişkeninin kendisiyle çok yüksek korelasyona sahiptir ve hatta ondan türetilmiştir. Yani, eğer bir müşterinin Churn Reason'ı varsa, o müşteri zaten churn etmiştir. Modelin, bir müşterinin churn edip etmediğini tahmin etmesini istiyorsanız, bu "gelecekten bilgi sızdıran" (data leakage) sütunlardır. Model, bu sütunlara bakarak churn edip etmediğini %100 doğru tahmin edebilir, ancak bu gerçekçi değildir çünkü churn nedenini henüz bilmiyorsunuzdur. Bu sütunları (Churn Reason, Churn Category) model eğitiminden önce tamamen silmek, modelinizin daha gerçekçi ve kullanılabilir olmasını sağlar.

2. Offer
Offer sütunu da verisetinin yarısından fazlasında boş değerlere sahiptir, bu büyük bir orandır. Bu sütun, müşteriye yapılan bir teklifi (örneğin, "Offer A", "Offer B" veya "No Offer") gösteriyor olabilir. Eksik değerler büyük ihtimalle müşteriye herhangi bir teklif yapılmadığı anlamına gelir.

Müdahale: Eksik değerleri "No Offer" gibi yeni bir kategori ile doldurmak en yaygın ve mantıklı yaklaşımdır. Bu, modelin "teklif almayan" müşteriler grubunu da öğrenmesini sağlar.

3. Internet Type
Internet Type, diğer sütunlara göre daha az eksik değer içeriyor. Bu sütun, müşterinin internet hizmeti türünü belirtiyor olabilir. Eksik değerler büyük ihtimalle müşterinin hiç internet hizmeti almadığı anlamına gelir. Internet Service sütunundaki "No" değerlerinin Internet Type değerinin noş olması beklenen bir durumdur.

Müdahale: Tıpkı Offer sütununda olduğu gibi, bu eksik değerleri "No Internet Service" gibi yeni bir kategori ile doldurmak mantıklıdır.

Tüm bu aşamalardan önce varsayılan durumların doğrulaması yapılmalı, sonra belirlenen eksik değerle başa çıkma stratejisi uygulanmalı ve yapılan işlem tekrar kontrol edilmelidir.

#### Demografik Özellikler

```python
# Teker teker görselleştirmek yerine benzer veri alanlarını birlikte alt grafikler olarak görselleştirmek için liste oluşturuyoruz
demographic_features = ['Gender', 'Under 30', 'Married', 'Senior Citizen', 'Dependents', 'Referred a Friend']

# Tek bir figure üzerinde birden fazla grafik göstermek için bu yapıyı kullanıyoruz
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
# 2D array'i 1D array'e dönüştürür, döngüde kolaylık sağlar
axes = axes.flatten()

# Listedeki her eleman için Churn Label ile bir count sütun grafik oluşturuyoruz
for i, col in enumerate(demographic_features):
    sns.countplot(data=df, x=col, ax=axes[i], hue='Churn Label', palette='RdPu')
    axes[i].set_title(f'{col} and Customer Churn Relationship', fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel('Number of Customers', fontsize=10)
    
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

##### Nasıl Değerlendirilir?
- Mean ve Median - 50%: Bunlar birbirine yakın mı? Eğer çok farklıysa, dağılımda çarpıklık (skewness) veya aykırı değerler olabilir.
  - Mean ve median arasındaki büyük farklar, veri setinde aşırı uç değerlerin (aykırı değerler) varlığına da işaret edebilir. Özellikle, Max değerlerinin 75% (3. çeyreklik) değerinden çok daha uzakta olması bu durumu destekler.
  - Mean > Median (Ortalama > Medyan): Sağa Çarpık (Right-Skewed) Dağılım, dağılımın sağa doğru "kuyruklu" olduğu anlamına gelir. Yani, verilerin büyük çoğunluğu düşük değerlerde toplanırken, bir kısım yüksek değerler (aykırı değerler veya sadece daha az sayıda yüksek gözlem) ortalamayı medyanın üzerine çekmektedir.
- Standart Sapma (std): Veri noktalarının ortalamadan ne kadar yayıldığını gösterir.
- Min ve Max değerler: Beklenen aralıkta mı? Çok uç değerler var mı?
- 25%, 50%, 75%: Verinin nasıl dağıldığı hakkında iyi bir fikir verir.

##### Değerlendirme Çıkarımları
- Total charges, number of referrals, total refunds, total extra data charges, total long distance charges, total revenue değerlerinde mean ve median değerleri arasında yüksek farklılıklar var.
- Total Charges, Total Refunds, Total Extra Data Charges, Total Long Distance Charges, Total Revenue değerlerinde "mean > median" durumu görülmektedir. Dağılımda çarpıklık olduğu söylenebilir. Ek olarak, bu alanların  Max - 75% karşılaştırılmasına bakıldığında, büyük farkları görmemiz aykırı değerlerin varlığını işaret eder.

```python
# Zip Code, Latitude, Longitude gibi alanlar istatistiksel bir bilgi vermeyeceği için numeric_cols listesinden çıkarılacaklar listesi oluşturuyoruz.
columns_to_remove = ['Zip Code', 'Latitude', 'Longitude']

# numeric_cols listesinden columns_to_remove listesini çıkarıyoruz
# Bu yöntem, numeric_cols listesindeki elemanları belirli bir koşula (columns_to_remove) göre filtreleyerek yeni bir liste oluşturur. 
# Orijinal listeyi değiştirmez, bu da çoğu zaman daha güvenlidir.
numeric_cols = [col for col in numeric_cols if col not in columns_to_remove]

# Oluşturulacak grafik sayısını, satır ve sütun şeklinde dinamik olarak belirliyoruz
num_plots = len(numeric_cols)
num_cols = 4
num_rows = (num_plots + num_cols - 1) // num_cols

# Dinamik olarak belirlediğimiz satır ve sütun sayılarıyla figure değişkenini oluşturuyoruz
figure, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(num_cols * 5, num_rows * 4))
axes = axes.flatten()

# numeric_cols listesindeki her alan için histogram-yoğunluk grafiği oluşturuyoruz
for i, col in enumerate(numeric_cols):
    sns.histplot(x=col, data=df, ax=axes[i], bins=30, kde=True, color='violet', edgecolor='darkviolet', alpha=0.7)
    axes[i].set_title(f'{col} Distribution', fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel('Number of Customers', fontsize=10)
    
# Grafik sayısını dinmaik olarak oluşturduğumuz için, fazla alt grafikleri kapatıyoruz (eğer num_plots, num_rows * num_cols'tan az ise)
for j in range(num_plots, len(axes)):
    figure.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
