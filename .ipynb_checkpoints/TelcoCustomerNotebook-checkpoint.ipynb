```python
# info() fonksiyonuyla da alınabilen bilgileri, aşağıdaki yöntemle, tablo halinde daha okunaklı şekilde gösterebiliriz.

df_info = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes.values,
    "Missing Value Count": df.isnull().sum().values,
    "Missing Value Ratio (%)": np.round(df.isnull().mean().values * 100, 2),
    "Filled Value Count": df.notnull().sum().values
})

df_info = df_info.sort_values("Missing Value Count", ascending=False).reset_index(drop=True)

df_info
```

### Eksik Değerlerin İncelenmesi ve Müdahale

```python
missing_value_count = df.isnull().sum() # Her sütundaki eksik değer sayısını hesaplıyoruz.
missing_value_count = missing_value_count[missing_value_count > 0] # Sadece eksik değeri olan sütunları seçiyoruz

# Eğer hiç eksik değer yoksa hata vermemek için kontrol ediyoruz.
if not missing_value_count.empty:

    missing_value_ratio = missing_value_count / len(df) * 100
    
    # Sadece eksik veri içeren sütunları karşılaştırmak için yeni bir data frame oluşturuyoruz.
    missing_df = pd.DataFrame({
    "Column Name": missing_value_ratio.index,
    "Missing Value Ratio": missing_value_ratio.values
    })
    
    # Oranlara göre sıralıyoruz. (isteğe bağlı, daha düzenli görünüm sağlar)
    missing_df = missing_df.sort_values(by='Missing Value Ratio', ascending=False)

    # Oluşturduğumuz data frame ile eksik veri görselleşitrmesi yapıyoruz.
    plt.figure(figsize=(15,10))
    sns.catplot(x="Column Name", y="Missing Value Ratio", data=missing_df, hue="Column Name", legend=False, kind="bar", palette="RdPu")
    plt.title("Missing Value Ratio")
    plt.xlabel("Missing Value Columns")
    plt.ylabel("% Ratio")
    plt.xticks(rotation=45, ha='right') # Sütun isimleri çakışmasın diye döndür
    plt.tight_layout()
    plt.show()

else:
    print("There is no missing value in the data set.")
```

Churn Reason, Churn Category, Offer, Internet Type sütunlarında boş değerler olduğunu görüyoruz. Bu değişkenler kategorik değişkenlerdir. Eğer sayısal değişkenler olsaydı bir korelasyon analizi yapmamız gerekebilirdi. Doğrusal bir korelasyon olup olmadığını incelememiz gerekirdi.

1. Churn Reason ve Churn Category:
Churn Reason ve Churn Category sütunlarındaki boş değer oranı çok fazla gözüküyor. Bu boş değerler, eğer müşterinin terk etmediği yani Churn Label = "No" olan değerlerse, Churn Reason ve Churn Category alanlarının boş değer olması beklenen bir durumdur. Öncelikle bu durumu kontrol etmek gerekir.

    Müdahale: Bu sütunlar, Churn Label hedef değişkeninin kendisiyle çok yüksek korelasyona sahiptir ve hatta ondan türetilmiştir. Yani, eğer bir müşterinin Churn Reason'ı varsa, o müşteri zaten churn etmiştir. Modelin, bir müşterinin churn edip etmediğini tahmin etmesini istiyorsanız, bu "gelecekten bilgi sızdıran" (data leakage) sütunlardır. Model, bu sütunlara bakarak churn edip etmediğini %100 doğru tahmin edebilir, ancak bu gerçekçi değildir çünkü churn nedenini henüz bilmiyorsunuzdur. Bu sütunları (Churn Reason, Churn Category) model eğitiminden önce tamamen silmek, modelinizin daha gerçekçi ve kullanılabilir olmasını sağlar.

2. Offer:
Offer sütunu da verisetinin yarısından fazlasında boş değerlere sahiptir, bu büyük bir orandır. Bu sütun, müşteriye yapılan bir teklifi (örneğin, "Offer A", "Offer B" veya "No Offer") gösteriyor olabilir. Eksik değerler büyük ihtimalle müşteriye herhangi bir teklif yapılmadığı anlamına gelir.

    Müdahale: Eksik değerleri "No Offer" gibi yeni bir kategori ile doldurmak en yaygın ve mantıklı yaklaşımdır. Bu, modelin "teklif almayan" müşteriler grubunu da öğrenmesini sağlar.

3. Internet Type:
Internet Type, diğer sütunlara göre daha az eksik değer içeriyor. Bu sütun, müşterinin internet hizmeti türünü belirtiyor olabilir. Eksik değerler büyük ihtimalle müşterinin hiç internet hizmeti almadığı anlamına gelir. Internet Service sütunundaki "No" değerlerinin Internet Type değerinin noş olması beklenen bir durumdur.

    Müdahale: Tıpkı Offer sütununda olduğu gibi, bu eksik değerleri "No Internet Service" gibi yeni bir kategori ile doldurmak mantıklıdır.

Tüm bu aşamalardan önce varsayılan durumların doğrulaması yapılmalı, sonra belirlenen eksik değerle başa çıkma stratejisi uygulanmalı ve yapılan işlem tekrar kontrol edilmelidir.

##### Nasıl Değerlendirilir?
- Mean ve Median - 50%: Bunlar birbirine yakın mı? Eğer çok farklıysa, dağılımda çarpıklık (skewness) veya aykırı değerler olabilir.
  - Mean ve median arasındaki büyük farklar, veri setinde aşırı uç değerlerin (aykırı değerler) varlığına da işaret edebilir. Özellikle, Max değerlerinin 75% (3. çeyreklik) değerinden çok daha uzakta olması bu durumu destekler.
  - Mean > Median (Ortalama > Medyan): Sağa Çarpık (Right-Skewed) Dağılım, dağılımın sağa doğru "kuyruklu" olduğu anlamına gelir. Yani, verilerin büyük çoğunluğu düşük değerlerde toplanırken, bir kısım yüksek değerler (aykırı değerler veya sadece daha az sayıda yüksek gözlem) ortalamayı medyanın üzerine çekmektedir.
- Standart Sapma (std): Veri noktalarının ortalamadan ne kadar yayıldığını gösterir.
- Min ve Max değerler: Beklenen aralıkta mı? Çok uç değerler var mı?
- 25%, 50%, 75%: Verinin nasıl dağıldığı hakkında iyi bir fikir verir.

##### Değerlendirme Çıkarımları
- Total charges, number of referrals, total refunds, total extra data charges, total long distance charges, total revenue değerlerinde mean ve median değerleri arasında yüksek farklılıklar var.
- Total Charges, Total Refunds, Total Extra Data Charges, Total Long Distance Charges, Total Revenue değerlerinde "mean > median" durumu görülmektedir. Dağılımda çarpıklık olduğu söylenebilir. Ek olarak, bu alanların  Max - 75% karşılaştırılmasına bakıldığında, büyük farkları görmemiz aykırı değerlerin varlığını işaret eder.

```python
# Zip Code, Latitude, Longitude gibi alanlar istatistiksel bir bilgi vermeyeceği için numeric_cols listesinden çıkarılacaklar listesi oluşturuyoruz.
columns_to_remove = ['Zip Code', 'Latitude', 'Longitude']

# numeric_cols listesinden columns_to_remove listesini çıkarıyoruz
# Bu yöntem, numeric_cols listesindeki elemanları belirli bir koşula (columns_to_remove) göre filtreleyerek yeni bir liste oluşturur. 
# Orijinal listeyi değiştirmez, bu da çoğu zaman daha güvenlidir.
numeric_cols = [col for col in numeric_cols if col not in columns_to_remove]

# Oluşturulacak grafik sayısını, satır ve sütun şeklinde dinamik olarak belirliyoruz
num_plots = len(numeric_cols)
num_cols = 4
num_rows = (num_plots + num_cols - 1) // num_cols

# Dinamik olarak belirlediğimiz satır ve sütun sayılarıyla figure değişkenini oluşturuyoruz
figure, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(num_cols * 5, num_rows * 4))
axes = axes.flatten()

# numeric_cols listesindeki her alan için histogram-yoğunluk grafiği oluşturuyoruz
for i, col in enumerate(numeric_cols):
    sns.histplot(x=col, data=df, ax=axes[i], bins=30, kde=True, color='violet', edgecolor='darkviolet', alpha=0.7)
    axes[i].set_title(f'{col} Distribution', fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel('Number of Customers', fontsize=10)
    
# Grafik sayısını dinmaik olarak oluşturduğumuz için, fazla alt grafikleri kapatıyoruz (eğer num_plots, num_rows * num_cols'tan az ise)
for j in range(num_plots, len(axes)):
    figure.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

##### Sayısal Verilerin Görselleştirmelerinin Yorumlanması
1. Total Charges: Aşırı sağa çarpık bir dağılım. Medyan kutunun alt sınırına çok yakın. Yüksek toplam faturaları olan müşteriler, grafikte yukarıda dağınık noktalar olarak görünen birçok aykırı değer olarak işaretlenmiş. Bu değişken için logaritmik dönüşüm gibi bir normalleştirme işlemi, çoğu makine öğrenimi modeli için faydalı olacaktır.
  
2. Number of Referrals: 0 değerinde çok yoğun bir kümeleme var (medyan 0). Üstte, 0'dan farklı ve referans veren müşterilerin aykırı değer olarak göründüğünü gösteriyor. Bu değişkenin aşırı sağa çarpık olduğunu ve büyük çoğunluğun 0 referans verdiğini teyit ediyor. Bu, modelleme için ikili bir özellik veya log dönüşümü gibi teknikler gerektirebilir.
   
3. Total Refunds: Yine "0" değerinde çok büyük bir yığılma ve çok sayıda aykırı değer (geri ödeme alan müşteriler) var. Medyan "0"da. Number of Referrals gibi, bu da binary bir özellik olarak değerlendirilebilir ("İade Aldı mı?").

4. Total Extra Data Charges: "0"da yoğun yığılma ve üstte çok sayıda aykırı değer. Müşterilerin büyük çoğunluğu ekstra veri ücreti ödemiyor. Binary özellik veya dönüşüm.

5. Total Long Distance Charges: Sağa çarpık dağılım, ve yüksek uzun mesafe ücreti ödeyen müşterilerin aykırı değer olarak göründüğü noktalar var.

6. Total Revenue: Total Charges'a benzer şekilde ağır sağa çarpık bir dağılım ve çok sayıda yüksek gelirli aykırı değer. Logaritmik dönüşüm muhtemelen gerekli olacaktır.

7. Monthly Charge: Medyan kutunun ortasına yakın, ancak sağa doğru uzanan bir kuyruk var. Bazı müşterilerin yüksek aylık faturaları olduğunu gösteren aykırı değerler de mevcut.

8. Tenure in Months: Dağılımın nispeten geniş olduğunu ve aykırı değerlerin (0 veya çok yüksek değerler gibi) nadir olduğunu gösteriyor. Medyan ve kutunun konumu, müşteri ömrünün dağılımını gösteriyor.


```python
# Daha fazla incelenmesi gereken sayısal alanları yeni görselleştirmeler için liste haline getirdik
selected_numeric_cols = ['Total Charges', 'Number of Referrals', 'Total Refunds', 'Total Extra Data Charges', 'Total Long Distance Charges', 'Total Revenue', 'Monthly Charge', 'Tenure in Months']

figure, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.flatten()

for i, col in enumerate(selected_numeric_cols):
    sns.kdeplot(x=col, ax=axes[i], data=df, hue='Churn Label', fill=True, palette='cool')
    axes[i].set_title(f'{col} and Churn Relationship', fontsize=12)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('')

plt.tight_layout()
plt.show()
```

- Total Charges: Toplam fatura miktarı daha düşük olan müşterilerin (genellikle daha yeni müşteriler) churn etme eğiliminde olduğunu gösterir. Yani, müşteri ne kadar az toplam fatura ödemişse, o kadar churn etme olasılığı yüksektir. Bu da Tenure (abonelik süresi) ile güçlü bir korelasyonu işaret eder.
- Number of Referrals: Hiç referans vermeyen müşteriler, churn etme olasılığı daha yüksek bir grubu oluşturuyor. Müşteri referans verdikçe (yani Number of Referrals arttıkça), churn etme olasılığı azalıyor ve sadakat artıyor.
- Total Refunds: Hiç iade almayan müşterilerin churn oranı, iade alanlara göre daha yüksektir. Bu durum, şirket için bir uyarı işaretidir. İade alan müşterilerin sorunları çözülmüş ve bu nedenle sadık kalmış olabilirlerken, hiç iade almayan ama memnuniyetsiz olan müşteriler sessizce terk ediyor olabilir.
- Total Extra Data Charges: Ekstra veri hizmeti kullanmayan müşterilerin churn etme olasılığı daha yüksektir. Bu, ek hizmetlerden faydalanmanın müşteri sadakatini artırabileceğini düşündürüyor. Müşteriler, daha fazla hizmet kullandıkça (veya daha fazla fatura ödedikçe) sağlayıcıya daha fazla bağlı kalma eğiliminde olabilirler.
- Total Long Distance Charges: Daha yüksek uzun mesafe ücreti ödeyen müşterilerin (yani bu hizmeti daha aktif kullananların) daha az churn etme eğiliminde olduğunu gösterir. Uzun mesafeli arama hizmetini yoğun kullanan müşteriler, bu hizmetin kendileri için değerli olduğunu düşünüyor ve bu nedenle daha sadık kalıyor olabilirler.
- Total Revenue: Şirkete sağladığı toplam gelir düşük olan müşterilerin churn etme olasılığı daha yüksektir. Bu, genellikle daha yeni müşterilerin veya daha az hizmet kullananların churn riskini taşıdığını destekler.
- Monthly Charge: Aylık faturaları daha yüksek olan müşterilerin churn etme eğilimi daha fazladır. Bu, müşterilerin yüksek fiyatlandırmadan veya fiyat/performans dengesizliğinden memnuniyetsizlik duyduğunu ve bu durumun churn için önemli bir tetikleyici olabileceğini düşündürüyor.
- Tenure in Months: Daha yeni müşteriler (daha kısa abonelik süresi olanlar) churn etme olasılığı en yüksek olan gruptur. Bu, müşteri kazanımından sonraki ilk aylarda (özellikle ilk 1-2 yıl içinde) müşteri sadakatini sağlamanın kritik olduğunu vurgular. Şirket, bu "kritik erken dönem" müşterilerine özel stratejiler geliştirmelidir.

#### Korelasyon Analizi (Sayısal Değişkenler Arasında)

Korelasyon analizi, özellikle de sayısal değişkenler arasındaki ilişkileri anlamak ve çoklu doğrusallık (multicollinearity) sorunlarını tespit etmek için kritik öneme sahiptir. Çoklu doğrusallık, iki veya daha fazla bağımsız değişkenin birbirleriyle yüksek derecede ilişkili olduğu durumlarda ortaya çıkar ve bu durum, bazı makine öğrenimi modellerinin (özellikle lineer regresyon, lojistik regresyon gibi) performansını ve yorumlanabilirliğini olumsuz etkileyebilir.

```python
# Korelasyon analizine katmayacağımız bazı sütunları, numeric_cols dizisinden çıkarıyoruz.
# 'Zip Code', 'Latitude', 'Longitude' zaten daha önceden çıkarmıştık, 'CLTV' gelecekteki bir metrik, 
columns_to_exclude_from_corr = ['CLTV', 'Churn Score']

# Korelasyon matrisine dahil edilecek temiz sayısal sütunlar
numeric_cols_for_corr = [col for col in numeric_cols if col not in columns_to_exclude_from_corr]

fig = plt.figure(figsize=(12,10))
correlation_matrix = df[numeric_cols_for_corr].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='BuPu', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Between Numerical Columns', fontsize=16)
plt.show()
```

##### Korelasyon Matrisi (HeatMap) Nasıl Yorumlanmalı?
1. Yüksek Pozitif Korelasyon (0.7 - 1.0 arası):
İki değişken birlikte hareket ediyor demektir; biri arttığında diğeri de artar. Eğer iki özellik arasında çok yüksek bir pozitif korelasyon (örneğin 0.9'un üzeri) varsa, bu çoklu doğrusallık anlamına gelebilir. Bu durumda, modelinize her iki özelliği de dahil etmek yerine birini dışarıda bırakmayı veya ikisinden yeni bir özellik türetmeyi düşünebilirsiniz.
2. Yüksek Negatif Korelasyon (-0.7 ile -1.0 arası):
İki değişken zıt yönde hareket ediyor demektir; biri arttığında diğeri azalır.
3. Orta Düzey Korelasyon (0.3 - 0.7 veya -0.3 ile -0.7 arası):
Değişkenler arasında belirgin, ancak çok güçlü olmayan bir ilişki vardır. Bu tür ilişkiler genellikle model için faydalıdır.
4. Düşük Korelasyon (0.0 - 0.3 veya -0.0 ile -0.3 arası):
Değişkenler arasında zayıf veya hiç doğrusal ilişki yoktur. Bu değişkenler, diğerleriyle bağımsız bilgi sağlayabilirler.

##### Yapılabilecekler
- Yüksek oranda ilişkili (çoklu doğrusallık gösteren) özelliklerden birini elemek, modelin karmaşıklığını azaltabilir, eğitim süresini hızlandırabilir ve bazı modellerin kararlılığını artırabilir.
- Eğer çok sayıda yüksek korelasyonlu değişken varsa, PCA (Temel Bileşen Analizi) gibi boyut indirgeme teknikleri düşünülebilir.

Bu bilgilere göre oluşan heatmap incelendiğinde, incelenmesi gereken korelasyonlar şunlardır;

- Tenure in Months: Total Charges ve Total Revenue ile arasında yüksek pozitif korelasyon olduğu görülüyor. Korelasyon oranları birbiriyle neredeyse aynı. Tenure in Months müşterinin şirkette ne kadar süre bulunduğunu gösterdiğinden bu süre arttıkça ödedikleri toplan ücretin ve şirkete sağladıkları toplam gelirin artmış olması beklenen bir durumdur. 0.67 ile ortadan yükseğe doğru bir korelasyon bulunan Total Long Distance Charges da benzer bir durumdur.
- Monthly Charge: Total Charges ve Total Revenue değerleriyle orta-yüksek düzeyli bir korelasyon inceliyoruz. Müşterinin aylık faturası arttıkça ödedikleri toplan ücretin ve şirkete sağladıkları toplam gelirin artmış olması beklenen bir durumdur.
- Total Charges ve Total Revenue arasındaki 0.97 oranındaki yükzek pozitif korelasyon bu iki değerin neredeyse aynı sayılabilecek kadar benzer olduğuna işaret etmektedir.
- Total Long Distance Charges: Avg Monthly Long Distance Charges, Total Charges, Total Revenue değerleriyle orta-yüksek korelasyon olması beklenen bir durumdur.
- Bu değerler dışındaki korelasyonlarda genellikle düşük korelasyon izlenmektedir. Bu durum modele bağımsız bilgi sağlayabilecek alanlar olduğunu gösterir.
- Aralarında orta düzeyli korelasyon bulunan alanlar ise model için faydalıdır. Örneğin, Total Revenue-Total Long Distance Charges, Monthly Charge-Total Revenue, Tenure in Months-Total Long Distance Charges

### EDA Sonuçları: Müdahale Gerektiren Temel Alanlar

1. Eksik Veri Alanları (Churn Reason, Churn Category, Offer, Internet Type):

    Churn Reason, Churn Category, Offer, Internet Type sütunlarında boş değerler olduğu gözlenmiştir. Churn Reason ve Churn Category alanları Churn Label ile yüksek korelasyon göstermektedir. Bu durumda bahsedilen alanlar data leakage durumuna sebep olabileceğinden, veri setinden çıkarılmıştır. Offer sütunu teklif kategorilerinden oluşmaktadır. Boş olan satırlar teklif almayan müşterileri temsil etmektedir. Bu boş satırlar "No Offer" yeni kategorisi ile doldurularak, modelin "teklif almayan" müşteriler grubunu da öğrenmesi hedeflenmiştir. Benzer şekilde Internet Type sütunundaki boş satırlar da yeni "No Internet Service" kategorisi ile doldurulmuştur.

2. Yeni Müşterilerdeki Yüksek Churn Riski (Tenure in Months, Total Charges, Total Revenue):

    Abonelik süresi (Tenure in Months) düşük olan (özellikle ilk 20 ay), toplam faturaları ve toplam gelirleri düşük olan müşterilerin churn etme olasılığı belirgin şekilde daha yüksektir. Bu, müşteri ilişkisinin erken dönemlerinde özel ilgi gerektirdiğini göstermektedir.
   
    Müdahale: Bu metriklerin churn ile güçlü ters ilişkisi, modelde önemli göstergeler olmasını sağlar. Total Charges ve Total Revenue arasındaki aşırı yüksek korelasyon (0.97) nedeniyle çoklu doğrusallık yönetimi (birini eleme) gereklidir.

4. Belirli Hizmet ve Kontrat Tiplerindeki Churn Oranı (Contract, Internet Service, Monthly Charge):

    Aydan aya (Month-to-Month) sözleşmesi olan müşteriler ve Fiber Optik internet hizmeti kullanan müşteriler arasında churn oranları belirgin şekilde yüksektir. Ayrıca, aylık faturaları yüksek olan müşterilerde de churn eğilimi gözlemlenmiştir.

   Müdahale: Bu kategorik değişkenlerin ve Monthly Charge'ın churn üzerinde güçlü bir etkisi vardır. Contract, Internet Service gibi kategorik değişkenlerin modelde kullanılabilmesi için One-Hot Encoding gibi sayısallaştırma teknikleri uygulanmalıdır.

4. "0" Değerinde Yoğunlaşan Sayısal Özellikler (Number of Referrals, Total Refunds, Total Extra Data Charges):

    Bu değişkenlerde değerlerin büyük bir kısmı "0"dır ve dağılımları aşırı sağa çarpıktır.

    Müdahale: Bu alanlar, "referans verdi mi?", "iade aldı mı?" veya "ekstra veri kullandı mı?" gibi ikili (binary) bir bilgiyi daha iyi temsil edebilir. Bu nedenle, özellik mühendisliği ile ikili (binary) kategorik değişkenlere dönüştürülmeleri düşünülmelidir.

6. Çarpık Dağılımlı Diğer Sayısal Özellikler (Total Long Distance Charges, Avg Monthly GB Download):

    Total Long Distance Charges gibi diğer sayısal değişkenler de belirgin sağa çarpık dağılımlara sahiptir.

    Müdahale: Bu çarpık dağılımlar, modelin performansını olumsuz etkileyebileceğinden, logaritmik dönüşüm gibi veri dönüşümü teknikleri uygulanmalıdır.

6. Sayısal Özelliklerin Ölçek Farklılıkları:

    Tenure in Months (0-72), Monthly Charge (yaklaşık 18-120) ve Total Charges/Total Revenue (0-8000+) gibi sayısal özelliklerin değer aralıkları arasında büyük farklılıklar bulunmaktadır.

    Müdahale: Çoğu makine öğrenimi modelinin doğru çalışabilmesi için tüm sayısal özelliklerin benzer bir ölçeğe getirilmesi önemlidir. Bu nedenle, özellik ölçeklendirme (Min-Max Normalizasyon veya Standardizasyon) gereklidir.

### Kategorik Verileri Sayısallaştırma (Encoding)
Makine öğrenimi modelleri genellikle metin tabanlı verileri doğrudan işleyemez. Bu nedenle, Gender, Contract, Internet Service gibi kategorik sütunlardaki bilgileri sayısal bir formata dönüştürmemiz gerekir. Bu sürece "encoding" (kodlama) denir.

```python
# İkili kategorik değişkenleri grupladık. Bu gruba Label Encoding işlemi uygulayacağız.
binary_ordinal_cols = [
    'Gender', 'Under 30', 'Senior Citizen', 'Married', 'Dependents',
    'Referred a Friend', 'Phone Service', 'Multiple Lines', 'Internet Service',
    'Online Security', 'Online Backup', 'Device Protection Plan',
    'Premium Tech Support', 'Streaming TV', 'Streaming Music',
    'Streaming Movies', 'Unlimited Data', 'Paperless Billing', 'Churn Label'
]
```

```python
# Dizimizdeki bütün kategorik değişkenlerin ikili (binary) değere sahip olduğundan emin olduk. Hepsinin eşsiz değerlerini listeledik.
for i in (binary_ordinal_cols):
    print(df[i].unique())
```

```python
le = LabelEncoder()
```

```python
# binary_ordinal_cols dizisindeki değişkenlerin hepsine label encoding işlemi uyguladık. 0-1 değerleri atandı.
for col in binary_ordinal_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# Nominal (İkiden fazla eşssiz değere sahip olan) kategorik değişkenleri grupladık. Bu gruba One Hot Encoding işlemi uygulayacağız.
nominal_cols = ['City', 'Offer', 'Internet Type', 'Contract', 'Payment Method', 'Customer Status']
```

```python
# nominal_cols dizisindeki bütün sütunların nominal olduğunu doğruladık. 
for i in (nominal_cols):
    print(df[i].unique())
```

```python
# get_dummies(), belirtilen kategorik sütunları alır ve her bir benzersiz kategori için yeni ikili (0/1) sütunlar oluşturur. Bu şekilde OHE işlemi yapılmış olur.
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
```
